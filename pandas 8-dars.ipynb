```python
data = {
    'brand': ['BMW', 'Ford', 'Ford', 'Ford', 'Audi'],
    'Mashina': ['X5', 'Mustang', 'Mustang', 'Focus', 'A6'],
    'Narx': [50000, 40000, 45000, 25000, 65000],
    'year': [2019, 2017, 2020, 2018, 2019],
    'km_driven': [50000, 60000, 55000, 70000, 65000]
}

DF = pd.DataFrame(data)


ford_mustang = DF[(DF['brand'] == 'Ford') & (DF['Mashina'] == 'Mustang')]


mustang_2018_keyin = ford_mustang[ford_mustang['year'] > 2018]

ortacha_narx = mustang_2018_keyin['Narx'].mean()

print("2018-yildan keyin chiqqan Mustang mashinalarining o'rtacha narxi:")
print(ortacha_narx)
```
